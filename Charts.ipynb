```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Using MATLAB
plt.plot(x_axis, y_axis)
```

```python
# Using object oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Using object oriented interface (different method as above)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Using object oriented interface (different method as above)
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label = 'Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.ylim(0, 45)
plt.title('PyBer Fare by Month')
plt.grid()
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color='green', marker='D', linewidth=2, label='Boston')
ax.set_title('PyBer Fare by MOnth')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0,45)
ax.legend()
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for i in y_axis:
    y_axis_larger.append(i*3)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], color='blue', alpha=0.5, edgecolors='black', linewidth=2, label='Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.invert_yaxis()
ax.set_xlim(0,50)
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values, colors=colors)
plt.show()
```
